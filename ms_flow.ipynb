```python
import contextlib

from urllib import parse
from automate.eserv import config_factory

msal_app = config_factory().credentials.manager.get_credential('microsoft-outlook').msal()
```

```python
(auth_code_flow := msal_app.initiate_auth_code_flow(scopes=[]))
```

```python
LINK: str = '...'  # paste from auth flow

session: dict[str, object] = {}
with contextlib.suppress(ValueError):
    session = msal_app.acquire_token_by_auth_code_flow(
        auth_code_flow,
        {
            key: next(iter(ls))
            for key, ls in parse.parse_qs(LINK.rsplit('?', maxsplit=1)[-1]).items()
        },
    )

(flow := session.get('flow', session))
```
